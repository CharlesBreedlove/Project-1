```python
income = pd.read_csv("Resources/income.csv")

income
age = pd.read_csv("Resources/age.csv")
```

```python
clean_age_df = pd.DataFrame()
clean_age_df["City-State"] = age["Geography"]
clean_age_df["Total Population"] = age["Total; Estimate; Total population"]
clean_age_df["Millennial Population"] = age["Total; Estimate; AGE - 20 to 24 years"] + age["Total; Estimate; AGE - 25 to 29 years"] + age["Total; Estimate; AGE - 30 to 34 years"] + age["Total; Estimate; AGE - 30 to 34 years"] + age["Total; Estimate; AGE - 35 to 39 years"]
clean_age_df["Millennial Income"] = income["Median income (dollars); Estimate; HOUSEHOLD INCOME BY AGE OF HOUSEHOLDER - 25 to 44 years"].map(
    "${:,.2f}".format)
clean_age_df.head()
```

```python
# use census MSA lat/long in gAPI
# variables for x (lat) and y (long)
city_df = clean_age_df
keywords=["brew", "brewpub", "brewery"]
lat = []
long = []

#Format Everything
city_df["City-State"] = city_df["City-State"].str.replace("Metro Area","")
city_df["City-State"] = city_df["City-State"].str.replace(" ","+")

for row in range(0,len(city_df)):
    city = city_df["City-State"].loc[row]
    url_string_call =  f"https://maps.googleapis.com/maps/api/geocode/json?address={city}&key={gkey}"
    brewery_data = requests.get(url_string_call)
    test_var = brewery_data.json()
    target_city = test_var["results"][0]['geometry']['location']
    target_coords = f"{target_city['lat']},{target_city['lng']}"
    params = {
        "location": target_coords,
        "radius": 8000,
        "keyword": keywords,
        "key": gkey
    }
    # Build the endpoint URL (Checks all ice cream shops)
    base_url = "https://maps.googleapis.com/maps/api/place/radarsearch/json"

    # Run a request to endpoint and convert result to json
    brewery_data = requests.get(base_url, params=params)

    # Print the assembled URL
    #print(brewery_data.url)

    # Convert to JSON
    brewery_data = brewery_data.json()

    # Print the number of ice cream shops
    print(len(brewery_data["results"]))
    nu_of_relavent_results=len(brewery_data["results"])
    nu_of_relavent_results

    city = city_df["City-State"].loc[0]
#url_string_call =  f"https://maps.googleapis.com/maps/api/geocode/json?address=Austin,+TX&key={gkey}"
url_string_call =  f"https://maps.googleapis.com/maps/api/geocode/json?address={city}&key={gkey}"
#url_string_call

brewery_data = requests.get(url_string_call)

test_var = brewery_data.json()

test_var["results"][0]


target_city = test_var["results"][0]['geometry']['location']
target_city

# Target city
#target_city={"lat": 39.099727, "lng": -94.578567}
target_coords = f"{target_city['lat']},{target_city['lng']}"

params = {
    "location": target_coords,
    "radius": 8000,
    "keyword": keywords,
    "key": gkey
}

# Build the endpoint URL (Checks all ice cream shops)
base_url = "https://maps.googleapis.com/maps/api/place/radarsearch/json"

# Run a request to endpoint and convert result to json
brewery_data = requests.get(base_url, params=params)

# Print the assembled URL
print(brewery_data.url)

# Convert to JSON
brewery_data = brewery_data.json()

# Print the number of ice cream shops
print(len(brewery_data["results"]))
nu_of_relavent_results=len(brewery_data["results"])
nu_of_relavent_results
```
